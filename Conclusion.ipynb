Findings and Reccomendation

This assignment involved predicting customer satisfaction based on a six question survey.
    
Each question asked the customer to rank satisfaction regarding an aspect of service, on a scale from 1 to 5: 1 being the most dissatisfied, 5 being the most satisfied.

In addition, each customer was marked as either happy or unhappy overall with the service.
The dataset was small, only 126 surveys, but there were four classification algorithms that showed promise in prediction.

## Metric Optimized For
Because the primary concern was flagging unhappy customers, the prediction algorithms used were trained to minimize false positives; that is, unhappy customers being incorrectly classified as happy. This meant maximizing recall.

## Algorithms Used
The classification algorithms that showed initial promise in predicting customer happiness were the following: Decision Tree Classifier, SGD Classifier, SVC Classifier, and NuSVC Classifier.
The optimal parameters for each of these algorithms using the Grid Search algorithm from the python sklearn package. After optimization, the SVC and NuSVC Classifiers had each achieved a recall of 72%, albeit with different parameters, the SGD Classifier a recall of 78%, and the DecisionTree Classifier a recall of 76%. When combined into a voting ensemble, the recall increased to 86%.

### Methods
These algorithms were used in two different ways. First, they were used to create a voting ensemble to make predictions.
Second, two of the four were used in Recursive Feature Elimination in order to rank the importance of the six factors.
This was done in two ways. First, with Recursive Feature Elimination (RFE), where the six inputs were ranked from most to least important.
Second, with Recursive Feature Elimination With Cross-Validation (RFECV), which chose one or more features to keep, and ranked the remaining ones according to relevance. 

### RFE vs RFECV
In the case of RFECV, the algorithms optimized recall for the class of unhappy customers, and the following results were reached for the two algorithms used:
|Algorithm|Question 1|Question 2|Question 3|Question 4|Question 5|Question 6|
|---------|----------|----------|----------|----------|----------|----------|
|SGD      |False(1)  |False(4)  |False(3)  |False(2)  |True      |True      |
|DecisTree|True      |False(2)  |True      |False(1)  |True      |True      |

What the above table shows is that the SGD classifier kept questions 5 and 6 when optimizing for recall, and ranked the four dropped questions in order of relevence, question 1 being most relevant, question 2 being least. The Decision Tree Classifier, on the other hand, kept all questions except for questions 2 and 4 when optimizing. It then ranked the remaining two questions, although this ranking is likely not significant, as there wer only two questions dropped.
There was also the straight RFE case, in which all six questions were individually ranked from most predictive to least predictive. In this case, recall was not being optimized, but rather overall accuracy in prediction.

Note that RFE itself did not work to optimize recall, but just worked to maximize accurate predictions.

Note that, from averaging the numbers, we can see that questions 1, 5, and 6 appear to be the most important questions, at least in terms of ranking overall accuracy. Questions 2 and 4 were not selected by the RFECV test for either algorithm, and both have a low ranking (indicated by a high number). Questions 5 and 6 were the only questions used by both algorithms in maximizing recall, and both have a high ranking (indicated by a low number). Question 1 and 3 require more consideration. Both were used by only one algorithm (Decision Tree) in maximizing recall. However, we see that question 3 was given a low ranking in the RFE test, while question 1 was given a high ranking, an average equal to that of question 6. Even in the RFECV, where the SGD Classifier selected neither, it ranked question 1 at the highest level for not selected, whereas it ranked question 3 at a low level.

# Voting Ensemble and Hypothetical Data
In addition to the RFE and RFECV, I also created a voting ensemble among four classifiers in order to predict customer happiness.
The voting ensemble was not used to rank questions, but it was used to maximize recall for unhappy customers, achieving 86%.
After refining the voting ensemble, I created twelve hypothetical datasets, each with a maximum or minimum value for one of the questions.
For example, one hypothetical dataset had a 5 out of 5 for question 3, and all other questions the same as the actual data (called bad-3). Another hypothetical dataset had a 1 out of 5 for question 2, and all other questions the same as the actual data (called good-2).

Using the voting ensemble on these hypothetical datasets found some extreme behavior for question 1. For good-1 (5 out of 5 for 1, all others the same), the voting ensemble predicted close to 100% happy customers. For bad-1, the ensemble predicted exactly 100% unhappy customers. No other question generated this extreme behavior.
It should be noted that this was only looking at the extremes (1 out of 5 vs 5 out of 5) and so it should not be surprising that the results are different from the RFE and RFECV.

# Reccomendations
The questions, 1 to 6, were as follows:
1. My Order Was Delivered on Time
2. The Contents of My Order Were as Expected
3. I Ordered Everything I Wanted
4. I Paid a Good Price for My Order
5. I Am Satisfied with My Courier
6. The App Makes Ordering Easy

Based on the results, questions 1, 5, and 6 are the aspects that should be focused on for improvement. Not that the first question, order arriving on time, was especially predictive for customers who were extremenly happy, or extremely unhappy. This may indicate that specific orders were very late, which is intolerable to customers. This is the most important feature to remedy, as very low ranking for ordering on time are nearly 100% likely to be unhappy.
Presumably, all orders are intended to arrive on time, and without knowledge of what's causing late orders, one possibility might be offering a partial refund if an order is more than a certian amount of time late.

As far as questions 5 and 6, these too had predictive power in determining customer satisfaction. In particular, satisfaction with the courier was important, indicating that courtesy is likely an issue for many customers. Courtesy training for employees is something that therefore should likely be improved.

Finally, satisfaction with the app was something that seemed to have an impact on customer happiness. Frustration with ordering from an interface is likely to taint customer satisfaction, as time wasted trying to figure out how to order something will give customers a bad review of the company.
While I do not have any knowledge of the app, a review of it might be in order, but a short term fix may simply be to have a hotline with someone trained to help customers with the app, or improving the instructions.
