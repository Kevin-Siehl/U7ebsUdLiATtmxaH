```python
import pandas as pd

from sklearn.model_selection import train_test_split, GridSearchCV
#from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import NuSVC
#from sklearn.neighbors import KNeighborsClassifier
#from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import recall_score, classification_report, confusion_matrix

import random
import math

from sklearn.metrics import make_scorer, recall_score
```

```python
# Load data
data = pd.read_csv("ACME-HappinessSurvey2020.csv")
#data = pd.read_csv("Inverted-Survey.csv")

# Separate features and target
X = data.drop("Y", axis=1)
y = data["Y"]
```

```python
test_split_seed = 1234 #random.randrange(1000,9999)
#print(test_split_seed)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=test_split_seed
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

recall_cl0_scr = make_scorer(recall_score, pos_label=0)
init_seed=1234
```

```python
dtl = NuSVC(random_state=init_seed, kernel='linear')#linear, poly, rbf, sigmoid

param_gridl = {
    'nu': [0.1, 0.25, 0.5],
    'class_weight': [None, 'balanced'],
    #'gamma': ['auto', 'scale', 0.1], #only matters for kernel=rbf,poly,sigmoid
    #'coef0': [0.0, 1.0], #only matters for kernel=poly,sigmoid
    #'degree': [1,2,3,4], #only for kernel=poly
    'shrinking': [True, False]
}

gridl = GridSearchCV(
    dtl,
    param_gridl,
    scoring= recall_cl0_scr,
    cv=5
)

gridl.fit(X_train_scaled, y_train)#5678->nu=0.25; 4 10 4 8
#                                  1234->nu-0.25; 4 10 4 8
```

```python
# Make predictions
y_predl = gridl.predict(X_test_scaled) #0.29

# Evaluate
print("recall:", recall_score(y_test, y_predl, pos_label=0))
#recall.append(recall_score(y_test, y_pred, pos_label=0))

print("\nClassification Report:")
print(classification_report(y_test, y_predl))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_predl))
#print(test_split_seed)
```

```python
dtp = NuSVC(random_state=init_seed, kernel='poly')#linear, poly, rbf, sigmoid

param_gridp = {
    'nu': [0.1, 0.25, 0.5],
    'class_weight': [None, 'balanced'],
    'gamma': ['scale', 'auto', 0.01, 0.1], #gamma_vars, #only matters for kernel=rbf,poly,sigmoid
    'coef0': [0.0, 0.5, 1.0], #only matters for kernel=poly,sigmoid
    'degree': [1, 2, 3, 4], #only for kernel=poly
    'shrinking': [True, False]
}

gridp = GridSearchCV(
    dtp,
    param_gridp,
    scoring= recall_cl0_scr,
    cv=5
)

gridp.fit(X_train_scaled, y_train)#5678->nu=0.1; degree=1; 2 12 3 9
#                                  1234->nu=0.1; degree=1; 2 12 3 9
```

```python
# Make predictions
y_predp = gridp.predict(X_test_scaled) #0.57 -> 0.14 but avoided error with unchosen degree=3,4 (degree=1 both cases)

# Evaluate
print("recall:", recall_score(y_test, y_predp, pos_label=0))
#recall.append(recall_score(y_test, y_pred, pos_label=0))

print("\nClassification Report:")
print(classification_report(y_test, y_predp))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_predp))
#print(test_split_seed)
```

```python
dtr = NuSVC(random_state=init_seed, kernel='rbf')#linear, poly, rbf, sigmoid

param_gridr = {
    'nu': [0.1, 0.25, 0.5],
    'class_weight': [None, 'balanced'],
    'gamma': ['scale', 'auto', 0.01, 0.1, 1.0], #only matters for kernel=rbf,poly,sigmoid
    #'coef0': [0.0, 1.0], #only matters for kernel=poly,sigmoid
    #'degree': [1,2,3,4], #only for kernel=poly
    'shrinking': [True, False]
}

gridr = GridSearchCV(
    dtr,
    param_gridr,
    scoring= recall_cl0_scr,
    cv=5
)

gridr.fit(X_train_scaled, y_train)#5678->nu=0.1; 6 8 1 11
#                                  1234->nu=0.1; 6 8 1 11
```

```python
# Make predictions
y_predr = gridr.predict(X_test_scaled) #0.43

# Evaluate
print("recall:", recall_score(y_test, y_predr, pos_label=0))
#recall.append(recall_score(y_test, y_pred, pos_label=0))

print("\nClassification Report:")
print(classification_report(y_test, y_predr))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_predr))
#print(test_split_seed)
```
